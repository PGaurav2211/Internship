5) Please visit https://www.cnbc.com/world/?region=world and scrap-
 a) headings
b) date
c) News link

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.cnbc.com/world/?region=world'
```

```python
response = requests.get(url)
html_content = response.content
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
headings = soup.find_all('a', class_='LatestNews-headline')
```

```python
headlines_list = []
dates = []
links = []
```

```python
for heading in headings:
    headlines_list.append(heading.text.strip())
    
```

```python
for index, headline in enumerate(headlines_list):
    print(f"{index + 1}. {headline}")
```
